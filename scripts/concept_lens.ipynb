```python
from nnsight import LanguageModel
from concept_lens import ov_lens

# choose your model
model_str = 'meta-llama/Llama-2-7b-hf'
# model_str = 'meta-llama/Meta-Llama-3-8B'
# model_str = 'allenai/OLMo-2-1124-7B'
model = LanguageModel(model_str, device_map='cuda', dispatch=True)
```

```python
# concept lens examples in paper: 7b models 
ov_lens('he was a lifelong fan of the cardinals', model, k=80, head_ordering='concept')
ov_lens('the secret meeting of the cardinals', model, k=80, head_ordering='concept') 
ov_lens('in the morning air, she heard northern cardinals', model, k=80, head_ordering='concept')
```

```python
# token lens examples
ov_lens('in the morning air, she heard northern cardinals', model, head_ordering='token')
ov_lens('in the morning air, she heard northern cardinals', model, head_ordering='token', offset=-2)
ov_lens('in the morning air, she heard northern cardinals', model, head_ordering='token', offset=-3)
```
